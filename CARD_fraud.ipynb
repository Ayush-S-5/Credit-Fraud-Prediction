```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loan=pd.read_csv('fraudTest.csv')
```

```python
loan.head()
```

```python
loan.info()
```

```python
loan=loan.drop(['Unnamed: 0','trans_date_trans_time','first','last','gender','cc_num','city_pop','merch_lat','merch_long','lat','long','unix_time','zip','trans_num'],axis=1)
```

```python
loan=loan.drop(['street'],axis=1)
loan.head()
```

```python
loan.isnull().sum()
```

```python
loan=loan.dropna(how='all')
print(loan[loan['is_fraud']==0]['is_fraud'].count())
print(loan[loan['is_fraud']==1]['is_fraud'].count())
```

```python
loan.head()
```

```python
import datetime
def calc_age(age):
  today = datetime.datetime.today()
  return today.year-age.dt.year
```

```python
loan['dob']=pd.to_datetime(loan['dob'])
```

```python
loan['age']=calc_age(loan['dob'])
```

```python
loan.drop(['city','state','job','dob'],axis=1,inplace=True)
```

```python
cat=pd.get_dummies(loan['category'],drop_first=True)
```

```python
merc=pd.get_dummies(loan['merchant'],drop_first=True)
```

```python
finish=pd.concat([loan,cat,merc],axis=1)
```

```python
loan=finish.drop(['category','merchant'],axis=1)
```

Training

```python
from sklearn.model_selection import train_test_split
```

```python
X=loan.drop('is_fraud',axis=1)
y=loan['is_fraud']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

Model


```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier(criterion='entropy')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='entropy')
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=200)
```

```python
lr.fit(X_train,y_train)
```

```python
dc.fit(X_train,y_train)
```

```python
pred_lr=lr.predict(X_test)
```

```python
pred_dc=dc.predict(X_test)
```

Model accuracy

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_lr))
```

```python
print(classification_report(y_test,pred_dc))
```
